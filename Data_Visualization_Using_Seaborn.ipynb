```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("C:/Users/Dhruvi/Desktop/Imscit/sem-8/Machine Learning using Python[MLP] Practical/customer.csv")
df
```

```python
#Data Visualization using seaborn
# Line Plot
sns.lineplot(df['Age'],df['Tenure'])
```

```python
# Scatter Plot
sns.scatterplot(df['Age'],df['Tenure'],hue = df['Geography'])
```

```python
# Bar Plot
sns.barplot(df['Age'].head(5),df['Tenure'].head(10))
```

```python
# histogram
sns.distplot(df['Age'].head(10))
sns.distplot(df['Age'].head(30),kde = False, bins= 10)
```

```python
# Box PLot
sns.boxplot(df['Age'].head(10))
```

```python
# Violin Plot
sns.violinplot(df['Age'].head(10))
```

```python
# Swarm Plot
sns.swarmplot(df['Age'].head(10),df['Gender'].head(10))
```

```python
# Point Plot
sns.pointplot(df['Age'].head(10),df['Gender'].head(10))
```

```python
# Count Plot
sns.countplot(df['Gender'].head(10))
```
